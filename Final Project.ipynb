### Evaluation of Significance

In the present case, a suitable null hypothesis would be that there is no relationship between `accomodates` and `price`. If this were true, we would expect the observed slope of the line of best fit between those variables to be zero.

Therefore, we simulated this null hypothesis by using the permutation method. We permuted the output column (`price`). Then, we plotted the data points for this permuted data along with the linear regression line of best fit for the permuted data.

```python
np.random.seed(2950)
```

```python
def permute(input_array):
    # shuffle is inplace, so copy to preserve input
    permuted = input_array.copy().values #convert to numpy array, avoiding warning
    np.random.shuffle(permuted)
    return pd.Series(permuted) # convert back to pandas
```

```python
permute(df['price'])
```

```python
import seaborn as sns

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12,6), sharey=True)
ax1 = sns.regplot(x=df['accommodates'], y=df['price'], ax=ax1)
ax1.set(xlabel='accommodates', ylabel='price')
ax1.set_title('True data')

ax2 = sns.regplot(x=df['accommodates'], y=permute(df['price']), ax=ax2)
ax2.set(xlabel='accommodates', ylabel='permuted price')
ax2.set_title('Permuted data')
plt.show()
```

Permuted data had flatter slope (almost straight line) than the true data. Now, if we permute our `price` data and fit a linear model to it, we'll have the result of a trial under the null hypothesis (that there is no relationship between `accommodates` and `price`). We know that there's no relationship because we shuffled the `price` randomly, breaking any possible conection to `accommodates`.

Then, we used permutation to estimate the probability of observing a slope as extreme as, or more extreme than, the one we measure in our data. We performed 500 permutations of the output column (`price`), fitting in each case a new linear regression model and measuring the resulting slope. After that, we calculated how many of the permuted slopes were of greater magnitiude than the slope of the linear regression line from our observed data.

```python
np.random.seed(2950)
```

```python
model = LinearRegression().fit(df[["accommodates"]], df["price"])
```

```python
steeper_slopes = 0
n_permutations = 500
permuted_slopes = np.zeros(n_permutations)

for i in range(n_permutations):
    permuted_model = LinearRegression().fit(df[['accommodates']], permute(df['price']))
    
    permuted_slopes[i] = permuted_model.coef_[0]
    
    if np.abs(permuted_model.coef_[0]) > np.abs(model.coef_):
        steeper_slopes += 1

print('Number of the permuted slopes greater magnitiude than the slope of the linear regression line from our observed data:', steeper_slopes)
print('Simulated p value (Fraction of permuted slopes more extreme than observed slopes):', {steeper_slopes / n_permutations})
```

**To evaluate the significance of our data analysis, we conducted a hypothesis test. The null hypothesis would be that there is no relationship between `accomodates` and `price`. If this were true, we would expect the observed slope of the line of best fit between those variables to be zero. As we predicted in our hypothesis tests, there was no permuted slopes greater magnitude than the slope of the linear regression line from our observed data and our simulated p value was 0. We concluded that there is no relationship between accommodates and price.**

After simulating this null hypothesis using the permutation method, we also performed F-test to understand if the given set of predictor variables are significant in explaining the variance of the dependent variable.

```python
x = df[['accommodates', 'bedrooms',
       'beds', 'minimum_nights', 'availability_365',
       'number_of_reviews', 'review_scores_rating',
       'reviews_per_month']]
y = df['price']
```

```python
import statsmodels.api as sm
```

```python
X = sm.add_constant(x)
print(X)
```

```python
results = sm.OLS(y, X).fit()
A = np.identity(len(results.params))
A = A[1:,:]
print(results.f_test(A))
```

We get an F-critical value of 1.9384 which is much lower than our F-statistic score of 410.24. Since, F-statistic > F-critical, we reject the null hypothesis, which means that the independent variables are jointly significant in explaining the variance of the dependent variable.

```python
print(results.summary())
```

In this, we can check the p-value (listed as Prob F-statistic) in the summary to determine whether to reject or accept the null hypothesis. Here the p-value is the probability that the null hypothesis for the full model is true (i.e., that all of the regression coefficients are zero). Since the p-value is approximately zero, we reject the null hypothesis. In other words, there is evidence that suggests that there is a linear relationship between `price` and the independent variables.

The actual price is not the important factor to give satisfaction to the visitors. It is how much the visitors satisfied with the quality of the stay compare to the payments they made. This result closely connects to the real world where data cannot be interpreted just as numbers but also closely linked to the psychology of people.

### Data Limitations

- **Some reviews may be spam that is allowed or has not been filtered on Airbnb.** <br>
Spam reviews are hard to differentiate from real reviews and is diffcult to account for them in analysis.<br><br>
- **Listings are showed as an approximate location from `0-450 feet`.** <br>
Listings in the same building are anonymized by Airbnb individually, and therefore may appear "scattered" in the area surrounding the actual address. Listings that are also at border sections of the neighborhood are subject to being part of different neighborhoods at every data scraping as this is an Airbnb policy for user privacy before booking.<br><br>
- **There are listings where the price is valued astronomically or at zero.** <br>
There are listings in the dataset where a number of the price is at `0` or at the max price of `10,000`. This is due to the fact that some of them have not been properly set for hosting or just as a placeholder value for guests to negotiate with hosts.<br><br>
- **What are the biases in your data or assumptions of your analyses that specifically affect the conclusions youâ€™re able to draw?** <br>
The assumptions of our analysis are that the review scores are all equally weighed and valued for all listings. However, even with the review scores, we are unsure of how many users have rated the hosts for each listing. A listing with a score of `5` may truly only have `1` review.<br><br>
